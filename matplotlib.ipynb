```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

##### Basic graph

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y)
plt.show()
```

##### Adding a x and y label

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.show()
```

##### Resizing the font of the title

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y)
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.show()
```

##### changing the ticks to integers number only

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y)
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.show()
```

##### Adding a legend

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,label='Straight Line')
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```

##### Changing the color of the line

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,label='Straight Line',color='red')
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```

##### Changing the linewidth of the line

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,label='Straight Line',color='red',linewidth=12)
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```

###### Using a different marker

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,label='Straight Line',color='red',linewidth=1,marker='+',markersize=10)
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```

##### marker with different color

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,label='Straight Line',
         color='red',linewidth=1,marker='+',
         markersize=10,markeredgecolor='blue')
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```

##### Short hand notation in plot function

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,'+--r',label='Straight Line')
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```

##### Adding another line to the plot

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,'+--r',label='Straight Line')

x2 = np.arange(1,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,'b',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'b--')
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```

##### Resize the figure

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
#Resize the graph
plt.figure(figsize=(5,3))
plt.plot(x,y,'+--r',label='Straight Line')

x2 = np.arange(1,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,'b',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'b--')
plt.title('Our first title',fontdict = {'fontsize':20,'fontweight':'bold'})
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.xticks([1,2,3,4,5])
plt.yticks([6,7,8,9,10])
plt.legend() #we first specify the label in plot() function
plt.show()
```
