### checking for outliers in the dataframe

```python
x = HealthStrokeDataset[['bmi','age','avg_glucose_level']]
x.plot(kind='box')
plt.title('Distribution of BMI, Age, and Average Glucose Level')
plt.show()
```

##### From the above graph, it shows that we have outliers in bmi and average glucose level columns.

#### checking for columns with outliers and sum the outliers in each column.

```python
column  = ['bmi','age','avg_glucose_level']
# Compute IQR for each column
Q1 = HealthStrokeDataset[column].quantile(0.25)
Q3 = HealthStrokeDataset[column].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Find the sum and columns with outliers
outlier_cols = {}
for col in column:
    sum_of_outliers = ((HealthStrokeDataset[col] < lower_bound[col]) |
                (HealthStrokeDataset[col] > upper_bound[col])).sum()
    if sum_of_outliers > 0:
        outlier_cols[col] = int(sum_of_outliers)
print(outlier_cols)
```

```python
# Compute IQR for bmi column
Q1_bmi = HealthStrokeDataset['bmi'].quantile(0.25)
Q3_bmi = HealthStrokeDataset['bmi'].quantile(0.75)
IQR_bmi = Q3_bmi - Q1_bmi
lower_bound_bmi = Q1_bmi - 1.5 * IQR_bmi
upper_bound_bmi = Q3_bmi + 1.5 * IQR_bmi

# Compute IQR for avg_glucose_level column
Q1_gl = HealthStrokeDataset['avg_glucose_level'].quantile(0.25)
Q3_gl = HealthStrokeDataset['avg_glucose_level'].quantile(0.75)
IQR_gl = Q3_gl - Q1_gl
lower_bound_gl = Q1_gl - 1.5 * IQR_gl
upper_bound_gl = Q3_gl + 1.5 * IQR_gl


#Dealing with the outliers
bmi_without_outliers = HealthStrokeDataset[((HealthStrokeDataset['bmi'] >= lower_bound_bmi) |
                (HealthStrokeDataset['bmi'] <= upper_bound_bmi))]['bmi']
bmi_median = float(np.round(bmi_without_outliers.median(), 2))


glucose_without_outliers = HealthStrokeDataset[((HealthStrokeDataset['avg_glucose_level'] >= lower_bound_gl) |
                (HealthStrokeDataset['avg_glucose_level'] <= upper_bound_gl))]['avg_glucose_level']
glu_median = float(np.round(glucose_without_outliers.median(), 2))
```

```python
# Replace outliers with the median value
HealthStrokeDataset["bmi"] = np.where((HealthStrokeDataset['bmi'] < lower_bound_bmi) |
                (HealthStrokeDataset['bmi'] > upper_bound_bmi), bmi_median,HealthStrokeDataset['bmi'])

HealthStrokeDataset["avg_glucose_level"] = np.where((HealthStrokeDataset['avg_glucose_level'] < lower_bound_gl) |
                (HealthStrokeDataset['avg_glucose_level'] > upper_bound_gl), glu_median,HealthStrokeDataset['avg_glucose_level'])
```

```python
X = HealthStrokeDataset[['smoking_status','bmi','avg_glucose_level','Residence_type','work_type', 'gender','age', 'hypertension', 'heart_disease', 'ever_married']]
y = HealthStrokeDataset['stroke']
```

```python
le = LabelEncoder()
scaler = StandardScaler()
```

```python
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for col in categorical_cols:
     X.loc[:, col] = le.fit_transform(X[col])
```

```python
X.head()
```

```python
X.loc[:,['bmi', 'avg_glucose_level', 'age']] = scaler.fit_transform(X[['bmi', 'avg_glucose_level', 'age']])
```

```python
X.head().round(3)
```

### Principal component analysis

```python
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
# Define pipeline for classification
pipeline = Pipeline([
    ('classifier', KNeighborsClassifier())
])

# Define hyperparameters grid for classification
param_grid = [
    {
        'classifier': [KNeighborsClassifier()],
        'classifier__n_neighbors': [5, 2, 3, 4, 6, 7],
        'classifier__weights': ['uniform', 'distance'],
        'classifier__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
        'classifier__leaf_size': [20, 10, 15, 25, 35, 5],
        'classifier__p': [1, 2]
    },
    {
        'classifier': [GradientBoostingClassifier(random_state=42)],
        'classifier__n_estimators': [30, 70, 120, 180],
        'classifier__learning_rate': [0.03, 0.15, 0.25],
        'classifier__max_depth': [4, 5, 10],
        'classifier__max_features': ['sqrt', 'log2', None],
        'classifier__min_samples_leaf': [1, 2, 4],
        'classifier__min_samples_split': [2, 5, 10]
    },
    {
        'classifier': [DecisionTreeClassifier(random_state=42)],
        'classifier__splitter': ["best", "random"],
        'classifier__max_depth': [None, 5, 10, 20, 30],
        'classifier__min_samples_leaf': [1,2, 4, 7, 5],
        'classifier__min_samples_split': [2, 3, 4,6],
        'classifier__max_features': ['sqrt', 'log2', None],
        'classifier__class_weight': [None, 'balanced']
    }
]
```

```python
# Perform RandomizedSearchCV
random_search = RandomizedSearchCV(pipeline, 
                                 param_grid, 
                                 cv=5, 
                                 scoring='neg_mean_absolute_error', 
                                 n_jobs=-1,
                                 error_score='raise')
```

```python
random_search.fit(X_train_pca, y_train)
```

```python
# Print best parameters
print("Best Parameters:", random_search.best_params_)
```

```python
#select the best model
best_model = random_search.best_estimator_
best_model
```

```python
# Evaluate on test data
y_pred = best_model.predict(X_test_pca)
```
